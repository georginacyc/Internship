### Merging DataFrames
joining two dataframes/tables based on key(s)

[Write a Pandas program to join the two given dataframes along rows and merge with another dataframe along the common column id.](https://www.w3resource.com/python-exercises/pandas/joining-and-merging/pandas-joining-and-merging-dataframe-exercise-5.php)

```python
student_data1 = pd.DataFrame({
        'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
         'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin'], 
        'marks': [200, 210, 190, 222, 199]})

student_data2 = pd.DataFrame({
        'student_id': ['S4', 'S5', 'S6', 'S7', 'S8'],
        'name': ['Scarlette Fisher', 'Carla Williamson', 'Dante Morse', 'Kaiser William', 'Madeeha Preston'], 
        'marks': [201, 200, 198, 219, 201]})

exam_data = pd.DataFrame({
        'student_id': ['S1', 'S2', 'S3', 'S4', 'S5', 'S7', 'S8', 'S9', 'S10', 'S11', 'S12', 'S13'],
        'exam_id': [23, 45, 12, 67, 21, 55, 33, 14, 56, 83, 88, 12]})

print("Original DataFrames:")
print(student_data1)
print(student_data2)
print(exam_data)

joined = pd.concat([student_data1, student_data2])
merged = pd.merge(joined, exam_data, on="student_id")

print("\nJoined DataFrames:")
print(joined)

print("\nMerged DataFrames:")
print(merged)
```

### Multi-key Left/Right Join

[Write a Pandas program to join (left join) the two dataframes using keys from left dataframe only.](https://www.w3resource.com/python-exercises/pandas/joining-and-merging/pandas-joining-and-merging-dataframe-exercise-8.php)

```python
data1 = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'P': ['P0', 'P1', 'P2', 'P3'],
                     'Q': ['Q0', 'Q1', 'Q2', 'Q3']}) 
data2 = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'R': ['R0', 'R1', 'R2', 'R3'],
                      'S': ['S0', 'S1', 'S2', 'S3']})
print("Original DataFrames:")
print(data1)
print("--------------------")
print(data2)

print("\nMerged Data (keys from data1):")
print(pd.merge(data1, data2, on=["key1", "key2"], how="left"))

print("\nMerged Data (keys from data2):")
print(pd.merge(data2, data1, on=["key1", "key2"], how="left"))
```

```python
data1 = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'P': ['P0', 'P1', 'P2', 'P3'],
                     'Q': ['Q0', 'Q1', 'Q2', 'Q3']}) 
data2 = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'R': ['R0', 'R1', 'R2', 'R3'],
                      'S': ['S0', 'S1', 'S2', 'S3']})

print("Original DataFrames:")
print(data1)
print("--------------------")
print(data2)

print("\nMerged Data (keys from data2):")
print(pd.merge(data1, data2, on=["key1", "key2"], how="right"))

print("\nMerged Data (keys from data1):")
print(pd.merge(data2, data1, on=["key1", "key2"], how="right"))
```

### Patching Null Values With Another DataFrame

[Write a Pandas program to Combine two DataFrame objects by filling null values in one DataFrame with non-null values from other DataFrame.](https://www.w3resource.com/python-exercises/pandas/joining-and-merging/pandas-joining-and-merging-dataframe-exercise-15.php)

```python
df1 = pd.DataFrame({'A': [None, 0, None], 'B': [3, 4, 5]})
df2 = pd.DataFrame({'A': [1, 1, 3], 'B': [3, None, 3]})
print("Original DataFrames:")
print(df1)
print("--------------------")
print(df2)

patched1 = df1.combine_first(df2)
patched2 = df2.combine_first(df1)
print("Patched: ")
print(patched1)
print("--------------------")
print(patched2)
```
